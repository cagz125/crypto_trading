```python
#import q    #
#import zmq  #import socket package
import numpy as np
import pandas as pd
import datetime as dt

from urllib.request import Request, urlopen
import json
from pandas.io.json import json_normalize
import time



def request_to_df(request):
    response = urlopen(request)
    elevations = response.read()
    data = json.loads(elevations)
    df= json_normalize(data['result'])
    return df


cnt=0
while cnt < 5:
    cnt+= 1
    print(cnt)
    if cnt == 1:
        df = request_to_df('https://bittrex.com/api/v1.1/public/getmarketsummaries')
        timestamp = dt.datetime.now()
        vt_ts = pd.to_datetime(np.repeat(timestamp,len(df)))
        vt_index = df['MarketName']
        df.index = [vt_ts ,vt_index]
        df.index.names = ['timestamp_ix', 'mn_ix']

        df.to_hdf('bittrex.h5','bittrex_tb',format='table', mode = 'w')

    else:
        df = request_to_df('https://bittrex.com/api/v1.1/public/getmarketsummaries')
        timestamp = dt.datetime.now()
        vt_ts = pd.to_datetime(np.repeat(timestamp,len(df)))
        vt_index = df['MarketName']
        df.index = [vt_ts ,vt_index]
        df.index.names = ['timestamp_ix', 'mn_ix']

        df.to_hdf('bittrex.h5','bittrex_tb',format='table', mode ='a', append = True)

    time.sleep(5)


print (pd.read_hdf('bittrex.h5', 'bittrex_tb', where = ["mn_ix= 'BTC-VTC'"]))


```
